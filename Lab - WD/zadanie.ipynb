```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
plt.figure(figsize=(8, 6))


sns.scatterplot(x='total_bill', y='tip', data=tips, palette='mako')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.boxplot(x='day', y='total_bill', data=tips)
# Dodaj punkty do boxplota
sns.swarmplot(x='day', y='total_bill', data=tips, color='black', alpha=0.6)
plt.title('Boxplot with Swarmplot')
plt.show()
```

```python
iris = sns.load_dataset("iris")
# Stwrz par plot z markerami i palet
sns.pairplot(iris, hue='species', markers=['o', 's', 'D'], palette='rocket')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')

g = sns.FacetGrid(titanic, col='survived')
g.map(plt.hist, 'age', bins=20, color='skyblue', edgecolor='black')

plt.show()
```
