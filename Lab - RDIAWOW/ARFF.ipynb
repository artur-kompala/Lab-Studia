# <center>Wczytywanie pliku ARFF</center>

```python
from scipy.io import arff
import pandas as pd

# Wczytanie pliku ARFF
data, meta = arff.loadarff('weather.nominal.arff')

# Przekonwertowanie na DataFrame
df = pd.DataFrame(data)

# Przekonwertuj kolumny typu bytes na stringi
for column in df.columns:
    if df[column].dtype == object:  # Sprawdzenie typu object
        if isinstance(df[column][0], bytes):  # Sprawdzenie, czy wartości są typu bytes
            df[column] = df[column].apply(lambda x: x.decode('utf-8'))

display(df)
```

# <center>Zapis do pliku ARFF</center>

```python
import arff  # To jest moduł z biblioteki liac-arff

# Tworzenie listy atrybutów automatycznie na podstawie danych w DataFrame
attributes = []
for column in df.columns:
    unique_values = df[column].unique()
    if df[column].dtype == object:
        # Dodanie atrybutu z listą unikalnych wartości dla kolumny typu object (kategoria)
        attributes.append((column, unique_values.tolist()))
    else:
        # Dodanie atrybutu typu NUMERIC dla kolumny liczbowej
        attributes.append((column, 'NUMERIC'))

# Struktura ARFF
arff_data = {
    'relation': 'dynamic_data',
    'attributes': attributes,
    'data': df.values.tolist()
}

# Zapis do pliku ARFF
with open('dynamic_data.arff', 'w') as f:
    arff.dump(arff_data, f)
```
