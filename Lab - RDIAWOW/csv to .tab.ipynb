```python
import re
def csv_to_rses_tab(filepath):
    with open(filepath, "r") as f1:
        # Odczytaj nagłówki
        header = f1.readline().strip().split(',')

        header = [x.replace(' ', '_') for x in header]

        # Tworzenie ścieżki do pliku .tab
        tab_filepath = filepath[:-4] + '.tab'

        with open(tab_filepath, "w") as f2:
            # Zapisz nagłówek tabeli
            f2.write(f"TABLE {filepath[:-4]}\n")
            f2.write(f"ATTRIBUTES {len(header)}\n")

            # Zapisz atrybuty (zakładam, że ostatni atrybut to klasa)
            for i, column in enumerate(header):
                if i == 0: #len(header) - 1:
                    f2.write(f"{column} symbolic\n")  # Klasa (np. disease)
                else:
                    f2.write(f"{column} numeric 0\n")  # Pozostałe atrybuty jako numeryczne

            # Ilość obiektów (wczytaj wszystkie wiersze)
            lines = f1.readlines()
            f2.write(f"OBJECTS {len(lines)}\n")

            # Zapisz dane, zastępując MISSING w CSV na 'MISSING'
            for line in lines:
                line = line.strip().replace(',', ' ')
                line = line.replace('MISSING', 'MISSING')  # Przykład, jeśli w CSV jest MISSING
                line = re.sub(r'\bTrue\b', '1', line, flags=re.IGNORECASE)
                line = re.sub(r'\bFalse\b', '0', line, flags=re.IGNORECASE)
                f2.write(line + '\n')

    print(f'Plik {tab_filepath} został utworzony.')

# Przykład użycia
csv_to_rses_tab('nowy.csv')
```
